```python
idvalues = np.unique(onn["H_ID"])
idvector = list(idvalues)
#len(idvector)
print(idvector)
```

```python
grouped = onn.groupby("H_ID")
```

```python
len(grouped)
```

```python
colnames = ["ID", "Section", "Type", "Length (km)", "ADT", "All injury acc", "Head-on", "VRU", "Other"]
df = pd.DataFrame(columns = colnames)
df["ID"] = idvector  
```

```python
# Create empty vectors. These will be filled with accidents of specified type and added to final dataframe
OtherVectorFinal = []
VRUVectorFinal = []
HeadOnVectorFinal = []
AllInjuryFinal = []

# Create empty vectors
WMEAN_KVL = []
pitklink_p = []
KATULUOKKA = []
distance = []
KATU_left = []

columnlist = ["distance", "WMEAN_KVL", "pitklink_p", "KATULUOKKA", "KATU_left"]
columnlistobj = [distance, WMEAN_KVL, pitklink_p, KATULUOKKA, KATU_left]

# Iterate over groups (road IDs)
for i in idvector:
    
    # reset vectors
    OtherVector = []
    VRUVector = []
    HeadOnVector = []
    
    # access road's accident types
    grp = grouped.get_group(i)
    grp = grp.reset_index()
    grp = grp[["olk2", "KATU_left", "WMEAN_KVL", "pitklink_p", "distance", "KATULUOKKA"]] # Insert colnames with road name + ADT stuff here
    
    #add necessary columns into a list of lists
    for x in range(0, len(columnlist)):
        columnlistobj[x].append(grp[columnlist[x]].iloc[0])

    # save road's accident types to vectors
    occurrencesOne = np.count_nonzero(grp["olk2"] == 1)
    HeadOnVector.append(occurrencesOne)
    occurrencesTwo = np.count_nonzero(grp["olk2"] == 2)
    VRUVector.append(occurrencesTwo)
    occurrencesThree = np.count_nonzero(grp["olk2"] == 3)
    OtherVector.append(occurrencesThree)
    
    # sum accident types for road being looped
    HeadOnRoad = sum(HeadOnVector)
    VRURoad = sum(VRUVector)
    OtherRoad = sum(OtherVector)
    AllInjury = (HeadOnRoad+VRURoad+OtherRoad)
    
    # Add sums to final vectors
    HeadOnVectorFinal.append(HeadOnRoad)
    VRUVectorFinal.append(VRURoad)
    OtherVectorFinal.append(OtherRoad)
    AllInjuryFinal.append(AllInjury)
    
#Add final vectors to df
df["Head-on"] = HeadOnVectorFinal
df["VRU"] = VRUVectorFinal
df["Other"] = OtherVectorFinal
df["All injury acc"] = AllInjuryFinal

for i in range(0, len(columnlist)):
    df[columnlist[i]] = columnlistobj[i]
    
#ADT and length
df["ADT_(Wmean)"] = df["WMEAN_KVL"]
df["Length (km)"] = df["pitklink_p"]/1000
df["Type"] = df["KATULUOKKA"]
df["Section"] = df["KATU_left"]
df = df.drop(["WMEAN_KVL", "KATU_left", "KATULUOKKA", "distance", "pitklink_p"], axis=1)


df
```
